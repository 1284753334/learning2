```python
'''
全球食品添加剂用量分析
1、pandas加载数据
2、按国家分组统计食品添加剂的平均用量
3、柱形图呈现食品添加剂平均用量最多的前十个国家
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#csv数据路径
csv_path = './data/FoodFacts.csv'
# 读取csv数据
df_obj = pd.read_csv(csv_path,usecols=['countries_en','additives_n','product_name'])
print(df_obj)
```

```python
# 数据清洗
# 去除缺失的数据
df_obj = df_obj.dropna()
print(df_obj)
```

```python
# 数据分析统计
data = df_obj['additives_n'].groupby(df_obj["countries_en"]).mean()
print(data)
```

```python
data = data.sort_values(ascending=False)
#print(data.head(10))
print(data[0:10])
```

```python
# 数据呈现
x = range(10)
print(x)
width = 0.75
plt.bar(x,data[0:10],width,color='b')
plt.savefig('./images/footfacts.png')
plt.show()
```

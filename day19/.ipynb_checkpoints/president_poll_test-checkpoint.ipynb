```python
# 需求：按照月份统计两位候选人民调的和，并且用可视化的方式呈现
# 实现的过程：
# 1、加载外部的csv数据，民调结束的时间，希拉里的民调数据，特朗普的民调数据
# 2、数据处理
#   1、月份做处理
#    a、日期处理成年份-月份格式
#    b、去重排序
#   2、两位候选人数据处理#@
#     a、缺失值处理
#     b、数据类型转换
# 3、数据分析
#     a、通过条件索引按月统计
# 4、数据展示
```

```python
import numpy as np
import datetime
```

```python
# 1、加载外部的csv数据
# 文件存储地址
filename='./data/presidential_polls.csv'
# 计算要加载列的索引
with open(filename,'r') as file:
    col_names_str = file.readline()[:-1]
print(col_names_str)
```

```python
col_names_list = col_names_str.split(',')
print(col_names_list)
```

```python
# 定义要使用的列
use_col_name_list = ['enddate','adjpoll_clinton','adjpoll_trump']
```

```python
# 计算要使用列的索引号
use_col_index_list = [col_names_list.index(use_col_name) for use_col_name in use_col_name_list]
print(use_col_index_list)
```

```python
data_array = np.loadtxt(filename,
                       delimiter=',',
                       skiprows=1,
                       dtype=str,
                       usecols=use_col_index_list)
print(data_array)
```

```python
# 数据处理
# 处理日期数据
enddate_lst = data_array[:,0].tolist()
print(enddate_lst)
```

```python
enddate_lst = [enddate.replace('-','/') for enddate in enddate_lst]
print(enddate_lst)
```

```python
print(datetime.datetime.strptime('10/27/2016','%m/%d/%Y'))
```

```python
enddate_lst = [datetime.datetime.strptime(enddate,'%m/%d/%Y') for enddate in enddate_lst]
print(enddate_lst)
```

```python
month_list = ["%d-%02d"%(dateobj.year,dateobj.month) for dateobj in enddate_lst]
print(month_list)
```

```python
# 去重排序
month_array = np.array(month_list)
months = np.unique(month_array)
print(months)
```

```python
# 处理希拉里的数据
adjpoll_clienton_data = data_array[:,1]
print(adjpoll_clienton_data)
```

```python
# 缺失值的处理
adjpoll_clienton_data = np.where(adjpoll_clienton_data == "","0",adjpoll_clienton_data)
# 数据类型的转换
adjpoll_clienton_data = adjpoll_clienton_data.astype(np.float64)
print(adjpoll_clienton_data)
```

```python
# 处理特朗普民调数据
adjpoll_trump_data = data_array[:,2]
print(adjpoll_trump_data)
```

```python
# 缺失值的处理
adjpoll_trump_data = np.where(adjpoll_trump_data=="",'0',adjpoll_trump_data)
# 数据类型的转换
adjpoll_trump_data = adjpoll_trump_data.astype(np.float64)
print(adjpoll_trump_data)
```

#数据分析  

```python
results = []
for month in months:
    bool_arr = month_array==month
    # 当月希拉里的民调数据
    adjpoll_clienton_month_data = adjpoll_clienton_data[bool_arr]
    adjpoll_clienton_month_sum =  adjpoll_clienton_month_data.sum()
    # 当月特朗普的民调数据
    adjpoll_trump_month_data = adjpoll_trump_data[bool_arr]
    adjpoll_trump_month_data_sum = adjpoll_trump_month_data.sum()
    results.append([month,adjpoll_clienton_month_sum,adjpoll_trump_month_data_sum])
```

```python
print(results)
```

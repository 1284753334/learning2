```python
import csv
import datetime
import random
import codecs
```

```python
csv_path = './data/data.csv'
with codecs.open(csv_path, 'w', encoding='utf-8') as file:
    wr = csv.writer(file)
    wr.writerow(['日期', '销量'])
    startdate = datetime.date(2018,1,1)
    # 模拟生成365个数据
    for i in range(365):
        if random.random() >0.5:
            amount = 400 + i + random.randrange(100)
        else:
            amount = 400 - i + random.randrange(100)
        wr.writerow([str(startdate),str(amount)])
        # 生成下一天的日期
        startdate = startdate + datetime.timedelta(days=1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 用pandas读取csv数据，创建DataFrame对象
df_obj = pd.read_csv('./data/data.csv')
print(df_obj)
```

```python
# 数据处理
df_obj = df_obj.dropna()
```

```python
# 生成折线图，反映商店全年每天营收的变化情况,保存图片first.png
plt.plot(df_obj['销量'])
plt.savefig('./images/first.png')
plt.show()
```

```python
# 按月进行统计，用柱形图呈现每月的营收情况,保存图片second.png
df_obj1 = df_obj[:]
df_obj1['month'] = df_obj1['日期'].map(lambda x:x[0:x.rindex('-')])
print(df_obj1)
```
